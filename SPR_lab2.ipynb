# Question 1: Basic Sinusoidal Signal


(a) Generate a basic sinusoidal signal in the time domain

(b) Plot the time-domain waveform

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
fs = 1000  # Sampling frequency in Hz
f = 5      # Frequency of the sine wave
t = np.arange(0, 1, 1/fs)  # Time vector from 0 to 1 second

# Generate the sine wave
x = np.sin(2 * np.pi * f * t)

# Plot the time-domain waveform
plt.figure()
plt.plot(t, x)
plt.title('Sine Wave (5Hz)')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

(c) Compute the Discrete-Time Fourier Transform (DTFT)

```python
# Compute the DTFT (approximation using high-resolution FFT)
n = len(x)
frequencies = np.linspace(-fs/2, fs/2, n)
X_dtft = np.fft.fftshift(np.fft.fft(x, n))

# Plot DTFT
plt.figure()
plt.plot(frequencies, np.abs(X_dtft))
plt.title('DTFT of Sine Wave (5Hz)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid()
plt.show()
```

(d) Compute the Discrete Fourier Transform (DFT)

```python
# Compute the DFT
X_dft = np.fft.fft(x)
frequencies_dft = np.fft.fftfreq(n, 1/fs)

# Plot DFT
plt.figure()
plt.stem(frequencies_dft[:n//2], np.abs(X_dft[:n//2]))
plt.title('DFT of Sine Wave (5Hz)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid()
plt.show()
```

# Question 4: Exponentially Decaying Signal

(a) Generate an exponentially decaying signal

(b) Plot the time-domain waveform


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
fs = 1000  # Sampling frequency
t = np.arange(0, 1, 1/fs)  # Time vector (1 second duration)
a = 5  # Decay constant

# Generate the exponentially decaying signal
x_exp = np.exp(-a * t)

# Plot the time-domain waveform
plt.figure()
plt.plot(t, x_exp)
plt.title('Exponentially Decaying Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

(c) Compute the Discrete-Time Fourier Transform (DTFT)

```python
# Compute the DTFT (using FFT approximation)
n = len(x_exp)
frequencies = np.linspace(-fs/2, fs/2, n)
X_exp_dtft = np.fft.fftshift(np.fft.fft(x_exp, n))

# Plot DTFT
plt.figure()
plt.plot(frequencies, np.abs(X_exp_dtft))
plt.title('DTFT of Exponentially Decaying Signal')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid()
plt.show()
```

(d) Compute the Discrete Fourier Transform (DFT)


```python
# Compute the DFT
X_exp_dft = np.fft.fft(x_exp)
frequencies_dft = np.fft.fftfreq(n, 1/fs)

# Plot DFT
plt.figure()
plt.stem(frequencies_dft[:n//2], np.abs(X_exp_dft[:n//2]))
plt.title('DFT of Exponentially Decaying Signal')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid()
plt.show()
```

Inference
The Fourier analysis of various signals provides insight into their frequency domain characteristics:

Sinusoidal signals have sharp peaks at their respective frequencies in both DTFT and DFT plots, demonstrating that they are composed of a single frequency component.
Composite signals show multiple peaks in the frequency domain, corresponding to each frequency component in the time-domain signal.
Exponentially decaying signals display a broad spectrum in the frequency domain, indicating that the signal has a wide range of frequency components due to its non-sinusoidal nature.
DTFT provides a continuous spectrum and offers a high-resolution view of the frequency components, whereas DFT, being discrete, gives a sampled version of the frequency spectrum with limited resolution depending on the number of points used.
